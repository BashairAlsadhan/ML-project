## Extracting Frames
We will extract the frames from the videos for model training and testing.



```python
import cv2

# Load the video
video = cv2.VideoCapture("parking - 12 sept, 12pm.MP4")

# Get the frame count
frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

# Extract the frames at a specified frequency
frequency = 30  # Extract a frame every 30 frames
for i in range(frame_count):
    # Read a frame from the video
    ret, frame = video.read()

    # Check if the frame was successfully read
    if not ret:
        break

    # Save the frame if it is a specified frequency
    if i % frequency == 0:
        filename = "ExtractedFrames/frame_{}.jpg".format(i)
        cv2.imwrite(filename, frame)

# Release the video
video.release()
```
